# Plotting

## per bead
* the top plot represents ramps (different cycles in different colors)
* the bottom row displays the estimated zmag closed of all beads (in black)
* the red spike represents the estimated zmag close for the currently displayed bead
* beads are reordered by zmag closed values

```python
%%opts Spikes [width=800]{+framewise} 
%%opts Spikes.Curr (line_color="red")
%%opts Curve [width=800] {+framewise} 
plotzzmag(data)
```

## Accross beads
* histzmagclose will display the histogram corresponding to distribution of zmag close

```python
%%opts Histogram [width=800]
histzmagclose(data,bins=20,range=(-0.6,-0.2))
```

## further note 
* If you need to access the data directly: 
    * data.zmagclose() will give you the dataframe (bead ids for rows , cycle ids for columns) of the **first closing event detected**
    * data.zmagclose(reverse_time=True), will give you the similar information but for the **last closing event**
